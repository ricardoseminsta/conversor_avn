```python
import pandas as pd
import glob

colunas=['id_seq', 'Identificação da Compra', 'Número do Item', 'Modalidade', 'Código do CATMAT', 'Item', 'Descrição',
         'Unidade de Fornecimento', 'Quantidade Ofertada', 'Valor Unitário', 'Fornecedor', 'Órgão', 'UASG - Unidade Gestora',
         'Objeto da compra', 'Valor proposto unitário', 'Forma da compra', 'Marca', 'Data resultado', 'CNPJ/CPF Fornecedor',
         'Porte Fornecedor', 'Orgão Superior', 'Número UASG', 'Ano compra', 'Painel', 'Data da Compra', 'Valor']

filenames = glob.glob("csv/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename, error_bad_lines=False, sep=';', index_col=False, header=0, names=colunas, dtype=str))


painel = pd.concat(dfs, ignore_index=True)
painel.to_excel('excel/painel.xlsx', index=False)
display(painel.info())
display(painel)
```

```python
mod_avn = pd.DataFrame()
mod_avn['id_compra'] = painel['Número UASG'] + "0" + painel['Painel'] + painel['Identificação da Compra'].str.replace('/','')
mod_avn['id_compra_item'] = mod_avn['id_compra'] + painel['Número do Item']
mod_avn['modalidade'] = painel['Modalidade']
mod_avn['catmat'] = painel['Código do CATMAT']
mod_avn['descricao'] = painel['Item']
mod_avn['desc_complet'] = painel['Descrição']
# mod_avn['qtd'] = painel['Quantidade Ofertada']
mod_avn['unidade'] = painel['Unidade de Fornecimento']
mod_avn['marca'] = painel['Marca']
mod_avn['metrica'] = painel['Forma da compra']
mod_avn['fornecedor'] = painel['Fornecedor']
mod_avn['uasg_comprador'] = painel['Número UASG']
mod_avn['nome_orgao'] = painel['UASG - Unidade Gestora'].str[9:]
mod_avn['orgao2_uasg'] = ''
mod_avn['orgao2_nome'] = painel['Orgão Superior']
mod_avn['orgao3_uasg'] = ''
mod_avn['orgao3_nome'] = painel['Órgão']
mod_avn['uf'] = ''
mod_avn['mes_compra'] = pd.to_datetime(painel['Data resultado'])
mod_avn['mes_compra'] = mod_avn['mes_compra'].dt.strftime('%b %Y')
mod_avn['valor'] = painel['Valor Unitário'].str.replace(',', '.')
mod_avn['valor'] = mod_avn['valor'].astype(float)
print(mod_avn.info())
```

```python
mod_avn.to_excel('cartao_memoria.xlsx', index=False)

# painel['Número UASG'].item
```
